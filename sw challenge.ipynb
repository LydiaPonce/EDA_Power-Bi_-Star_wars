# Planetas

```python
planets = pd.read_csv('planets.csv')
planets.head()
```

```python
planets.dropna(subset=['name']) #eliminan nan de nombre
```

```python
planets.rename(columns= {"name":"name_planets", },inplace=True)
```

```python
planets= planets.drop(['rotation_period','orbital_period', 'diameter', 'gravity', 'surface_water'], axis=1)
```

### Population

```python
pd.set_option('display.float_format', lambda x: '%.1f' % x) # quitar exponenciales
```

```python
population= planets['population'].replace((np.nan,None),inplace=True) #cambiar nan por none para que no afecten nan a las transformaciones
planets['population'].astype(int) #cambio a numero entero
```

```python
planets[["terrain","terrain_2", "terrain_3", "terrain_4"]] = planets["terrain"].str.split(pat=", ", n=-1, expand=True)
```

```python
planets.replace(np.nan,None)
```

```python
planets = planets.reindex(columns=['name_planets','climate','population','terrain','terrain_2','terrain_3', 'terrain_4'])
planets.head()
```

```python
planets.to_csv('planets_clean.csv')
```
